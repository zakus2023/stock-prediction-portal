## Subplots and legends

```python
days = [1,2,3,4,5]
original_price = [10,20,30,40,50]
predictedPrice = [8,21,25,28,38]
```

```python
plt.plot(days, original_price, "b-o", label="Original Price"); # b means blue color, -o show the data points, label="..." legend
plt.plot(days, predictedPrice, "r-o", label="Predicted Price");
plt.legend();
plt.title("plot with legends");
plt.xlabel("Weekdays");
plt.ylabel("Price");
```

## subplots

```python
# for original price
plt.subplot(3,1,2)
plt.plot(days, original_price, 'b-o', label="Original Price")
plt.legend();
plt.title("Original Price");
plt.xlabel("Weekdays");
plt.ylabel("Price");
```

```python
# for predicted price
plt.subplot(3,1,2)
plt.plot(days, predictedPrice, 'r-o', label="Predicted Price")
plt.legend();
plt.title("Predicted Price");
plt.xlabel("Weekdays");
plt.ylabel("Price");
```

## the difference between original price and predicted prices

```python
price_difference = [o - p for o, p in zip(original_price, predictedPrice)]
plt.subplot(3,1,3)
plt.bar(days, price_difference, color="purple", label="Price Difference")
plt.legend();
plt.title("Difference between original price and predicted price");
plt.xlabel("Weekdays");
plt.ylabel("Price Difference");
```

## saving the figure

```python
plt.plot(days, original_price, "b-o", label="Original Price"); # b means blue color, -o show the data points, label="..." legend
plt.plot(days, predictedPrice, "r-o", label="Predicted Price");
plt.legend();
plt.title("plot with legends");
plt.xlabel("Weekdays");
plt.ylabel("Price");
plt.savefig("Sample_prediction.png", dpi=300)
```
